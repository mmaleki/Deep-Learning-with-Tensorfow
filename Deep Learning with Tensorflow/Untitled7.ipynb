```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

```python
xs=[1,2,3,4]
ys=[2,1,4,5]
```

```python
plt.plot(xs,ys,'b*')
```

```python
def l(w0,w1):
    y=w0+np.dot(w1,xs)
    temp=(y-ys)**2
    loss=np.mean(temp)
    
    dtemp0=2*(y-ys)
    dw0=np.mean(dtemp0)
    
    dtemp1=2*(y-ys)*xs
    dw1=np.mean(dtemp1)
    
    return {'dw0':dw0,'dw1':dw1}
```

```python
l(2,3)
```

```python
def update(w0,w1,alpha):
    dl=l(w0,w1)
    dw0=dl['dw0']
    dw1=dl['dw1']
    w0=w0-alpha*dw0
    w1=w1-alpha*dw1
    return (w0,w1)
    
```

```python
w0=np.random.randn(1)[0]
w1=np.random.randn(1)[0]
ww0=[]
ww1=[]
for i in range(100):
    new_w=update(w0,w1,0.01)
    w0=new_w[0]
    w1=new_w[1]
    ww0.append(w0)
    ww1.append(w1)
    plt.plot(ww0,ww1,'bo-')
    
```

```python
t=np.linspace(0,5,20)
```

```python
plt.plot(xs,ys,'r*')
plt.plot(t,w0+w1*t)
```

```python
def l2(w0,w1):
    y=w0+np.dot(w1,xs)
    temp=(y-ys)**2
    loss=random.choice(temp)
    
    dtemp0=2*(y-ys)
    dw0=random.choice(dtemp0)
    
    dtemp1=2*(y-ys)*xs
    dw1=random.choice(dtemp1)
    
    return {'dw0':dw0,'dw1':dw1}
```

```python
l2(3,4)
```

```python
def update2(w0,w1,alpha):
    dl=l2(w0,w1)
    dw0=dl['dw0']
    dw1=dl['dw1']
    w0=w0-alpha*dw0
    w1=w1-alpha*dw1
    return (w0,w1)
```

```python
update2(3,4,0.2)
```

```python
w0=np.random.randn(1)[0]
w1=np.random.randn(1)[0]
ww0=[]
ww1=[]
for i in range(20):
    new_w=update2(w0,w1,0.01)
    w0=new_w[0]
    w1=new_w[1]
    ww0.append(w0)
    ww1.append(w1)
    plt.xlabel('w0')
    plt.ylabel('w1')
    plt.plot(ww0,ww1,'bo-')
```
